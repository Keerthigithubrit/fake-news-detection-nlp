# **Fake and Real News Detection-NLP Project**

This project implements a Fake News Detection system using Python, scikit-learn, and Natural Language Processing (NLP) techniques. The model classifies news articles as Fake or Real with high accuracy using Logistic Regression.

## **Overview**

Fake news has become a significant issue in the digital era. This project uses text classification techniques to identify fake news from real news based on textual content.

**Key steps:**

* Data loading and preprocessing
* Text cleaning (removing punctuation, numbers, stopwords, etc.)
* TF-IDF vectorization
* Logistic Regression model training
* Model evaluation
